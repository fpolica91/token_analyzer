```python
import pandas as pd
import sqlite3
import json

conn = sqlite3.connect('crypto_data.db')

# Read main market data
filename = 'reports/coinmarketcap.csv'
df_market_data = pd.read_csv(filename)
df_market_data.to_sql('market_data', conn, if_exists='replace', index=False)

# Process chart data
for _, item in df_market_data.iterrows():
    try:
        slug = item['slug']
        filename = f"reports/{slug}/chart_data.json"
        
        all_records = []
        with open(filename, 'r') as f:
            data = json.load(f)
            for timestamp, point in data['data']['points'].items():
                record = {
                    'timestamp': timestamp,
                    'slug': slug
                }
                
                # # Add all v values as separate columns
                for idx, val in enumerate(point.get('v', [])):
                    record[f'v{idx}'] = val
             
                
                # # Add all c values as separate columns    
                for idx, val in enumerate(point.get('c', [])):
                    record[f'c{idx}'] = val
              
                all_records.append(record)
        
        if all_records:
            df_chart = pd.DataFrame(all_records)
            df_chart.to_sql('chart_data', conn, if_exists='append', index=False)
            
    except Exception as e:
        print(f"Error processing {slug}: {e}")
        continue

conn.close()
```
